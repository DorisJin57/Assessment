```python
import pandas as pd
import numpy as np


data = {
    "city": ["Vancouver", "Toronto", "Vancouver", "Toronto", "Vancouver"],
    "price": [100, np.nan, 110, 130, 105],
    "type": ["A", "A", "B", "A", "A"]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby("city")["price"].mean().reset_index()
```

```python
df.groupby("city")["price"].count()
#city和price错位的原因是这里是city已经不是attribute name而是index
#加上.reset_index()可以转换成data form的格式
```

```python
df.groupby("city")["price"].agg(["mean", "count", "max"])
```

```python
df.groupby(["type", "city"])["price"].agg(["mean", "count", "max"])
```

```python
df.groupby(["city", "type"])["price"].mean()
```

```python
df.groupby(["city", "type"]).mean()
```

```python
df.groupby(["city", "type"]).size().reset_index(name="count")
```

```python
df.groupby(["city", "type"])["price"].size().reset_index(name="count")
```
